```python
# 1. 산술 연산자
a = 5
b = 2

print(a + b)   # 덧셈 : 7
print(a // b)  # 몫 :2
print(a % b)   # 나머지 : 1
print(5 ** 2)  # 거듭제곱 : 25
```

```python
# 2. 비교 연산자 : 연산(표현식)의 결과가 True 또는 False로 반환된다.
print(5 > 3)      # True
print(5 == 3)     # == : 같다 / False
print(5 != 3)     # != : 다르다 / True
```

```python
# 3. 논리 연산자
print(True and False)   # False
print(True or False)    # True
print(not True)         # False

print(10>5 and 10 == 7)   # True and False
```

```python
# 4. 멤버십 연산자 : A in B : B안에 A가 있는지 없는지 묻는것
print(3 in [1, 2, 3])       # True
a = [1, 2, 3]
print(4 in a)  # in : 포함됨 / False
print(4 not in a)   # not in : 포함되지 않음 / True
```

```python
# 식별 연산자 : 두 변수가 같은 객체를 참조하는지 확인
# is : 동일 객체 / is not : 다른 객체
a = [1, 2]
b = a 

a is b      # True / is : 객체 비교
a == b      # True / == : 객체의 값만 비교 

c = [1, 2]
d = [1, 2]

c is d      # False : 객체를 비교하기 때문에 False
c == d      # True : 값을 비교하기 때문에 [1,2] 는 같아서 True
```

```python
# 6. 삼항 연산자

score = 50
result = "합격" if score >= 60 else "불합격" 
print(result)

score = 70
result = "합격" if score >= 60 else "불합격" 
print(result)
```

```python
# 7. 대입 문

a = 10

# a = a+ 5와 같은 표현(편한거 사용하면 된다) 
a += 5
print(a)  # 15
```

```python
# 8. 월러스 연산자(대입 후 대입된 값을 반환하는 연산자) : python3.8 이후

# 파이썬에서는 다음과 같은 코드는 에러가 발생한다.
# 대입 연산자 우측에 표현(식)이 와야 하는데 우측 ()안에 대입문을 상용해서 에러 발생.
# = 는 연산자(operator)가 아니라 대입문(statement) 이기 때문이다.
# = : 대입문 → 값을 반환하지 않음 
# := : 대입 연산자 → 값을 반환함

​n = 10
result = (x = n>5)   # 대입하면서 값
print(result) # 에러 발생! (x = n>5)가 문이라서 반환값이 없음 (none)
```

```python
# := 월러스 연산자 사용, result = 우측 표현식 자리에 대입문이 아닌 대입연산자(:=)를 사용 
# 월러스 연산자는 왼쪽변수에 우측 값을 할당하고 할당된 값을 반환하는 대입연산자이다.

n = 10
result = (x := n>5)   # 대입하면서 값 반환
print(result)
```

```python
# 다른 언어에서는 다음과 같은 if문 표현식 자리에 =를  많이 사용하는데, 파이썬의 경우 표현식 자리에 =를 사용할 수 없기때문에 아래와 같은 표현은 에러가 발생한다.

# s변수의 문자열의 길이가 5보다 크면 문자열 길이를 출력
s = "python"
if (n = len(s)) > 5:
    print(n)
```

```python
# python 3.8 이전에는 위와 같은 표현은 다음과 같이 구현했다.

s = "python"
n = len(s)
if n > 5:
    print(n)
```

```python
# python 3.8에서 월러스 연산자가 추가된 후에는 다음과 같은 구현 가능

s = "python"
if (n := len(s)) > 5:
    print(n)
```

9. 연산자 규칙

​

1) 혼합된 연산자는 연산자 우선순위를 따른다. 하지만 긴 표현식에서는 표현식의 가독성에 도움이 되는 괄호()를 사용한다. 괄호는 실행의 순서를 강제하는 것뿐만 아니라 표현식의 가독성에 도움이 된다

​

2) 타입이 다른 숫자 타입끼리 연산하면 더 넓은 범위의 타입으로 자동 변환(Type Promotion) 한다.

int + int -> int 반환 

float + float -> float 반환

int + float -> float + float로 변환 후 연산 -> float 반환
2 + 2.0 = 4.0
