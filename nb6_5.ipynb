ETL에 자주사용되는 데이터타입은 엑셀같은 2차원 테이블 타입이다.
- Extract(추출), Transform(변환), Load(적재)

파이썬에는 2차원 테이블(행과 열) 전용 데이터 타입은 없지만, 보통 리스트 안에 사전(dict)을 넣는 형태로 표현합니다.

```python
students = [
    {"이름": "홍길동", "점수": 90},
    {"이름": "김철수", "점수": 85},
    {"이름": "이영희", "점수": 95}
]

print(students)
students
```

스키마 (구조)

- [] : list를 이용하여 여러 행(row)을 저장 
- {} : dict을 이용하여 한 행의 열(column) 데이터를 저장

```python
# 데이터 분석 실습
# [데이터 수집]
# 1) 데이터 준비

students = [
    {"이름": "김철수", "성적": 95, "주민번호": "980101-1234567"},
    {"이름": "이영희", "성적": 88, "주민번호": "010305-4234567"},
    {"이름": "박민수", "성적": None, "주민번호": "950720-1234567"},  # 결측치
    {"이름": "최지은", "성적": 77, "주민번호": "030215-4234567"},
    {"이름": "정우성", "성적": 105, "주민번호": "990101-1234567"},   # 이상치
    {"이름": "한지민", "성적": 82, "주민번호": "020912-4234567"},
    {"이름": "강호동", "성적": -10, "주민번호": "700511-1234567"},   # 이상치
    {"이름": "신세경", "성적": 91, "주민번호": "971224-2234567"},
    {"이름": "유재석", "성적": 85, "주민번호": "720814-1234567"},
    {"이름": "송혜교", "성적": None, "주민번호": "811122-2234567"}, # 결측치
    {"이름": "차은우", "성적": 79, "주민번호": "970330-1234567"},
] 

len(students) # 11행의 데이터 준비

# None = 결측치
# -10, 105 = 이상치 (우리가 설정해 나가야함)
```

```python
# [데이터 전처리]
# 2) 결측치 확인

missing_count = 0 # 결측지가 몇 개인지?

for student in students:
    if student["성적"] is None:
        missing_count += 1

print("결측치 개수:", missing_count)
```

```python
# 3) 결측치 처리(평균값 대체 : 어떤 값으로 대처하는지는 요구사항에 따라 다름)

scores = []

for student in students:
    if student["성적"] is not None and 0 <= score <= 100:
        scores.append(student["성적"]) 

avg = int(sum(scores) / len(scores))

for student in students:
    if student["성적"] is None:
        student["성적"] = avg

students #쥬피트 노트북에서는 print(students)로 출력하는것보다 변수이름으로 출력하는 것이 출력결과의 가독성이 좋음
```

```python
# 4) 이상치 탐지(정상 범위 : 0 ~ 100점)

for student in students:
    score = student["성적"]

    if score < 0 or score > 100:
        print(student)
```

```python
# 5) 이상치 처리(어떻게 처리할지는 요구사항에 따라 다름)

for student in students:

    if student["성적"] > 100:
        student["성적"] = 100

    elif student["성적"] < 0:
        student["성적"] = 0

students
```

```python
# 6) 성별 구하기(특성 추출)

for student in students:

    gender_code = student["주민번호"].split("-")[1][0] # 주민번호를 - 기준으로 나누고, 그 리스트의 [1]번째인 뒷자리 번호, 그 뒷자리 번호중 [0]번째 숫자 가져옴

    if gender_code in ["1", "3"]:
        student["성별"] = "남"
    else:
        student["성별"] = "여"

students
```

```python
# 7) 나이 구하기(특성 추출)

from datetime import datetime

current_year = datetime.now().year # 현재 날짜 데이터 중 연도 정보 사용

for student in students:
    rrn = student["주민번호"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["나이"] = current_year - birth_year

students
```

```python
# [EDA : 탐색적 데이터 분석]
# 8) 기술통계

scores = []

for student in students:
    scores.append(student["성적"])

print("개수:", len(scores))
print("합계:", sum(scores))
print(f"평균: {sum(scores) / len(scores):.2f}")
print("최대:", max(scores))
print("최소:", min(scores))
```

```python
# 9) 성별 평균 성적

male_sum = 0
male_count = 0

female_sum = 0
female_count = 0

for student in students:

    if student["성별"] == "남":
        male_sum += student["성적"]
        male_count += 1

    else:
        female_sum += student["성적"]
        female_count += 1

print(f"남 평균: {male_sum / male_count:.2f}")
print(f"여 평균: {female_sum / female_count:.2f}")
```

```python
# 10) 연령대별 평균 성적

age_group = {} # 연령대별 학생 목록 저장할 딕셔너리 생성. 예) {20 : [학생1], [학생2], 30 : [학생3], [학생4] ...}

for student in students:

    decade = (student["나이"] // 10) * 10 # // : 몫만 남기는 연산자

    if decade not in age_group:
        age_group[decade] = [] # 해당 연령대 key 생성 후 값으로 빈 리스트 연결

    age_group[decade].append(student["성적"]) # 빈 리스트에 성적 값 추가

print(age_group)

for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}대 평균: {avg:.2f}")
```
