```python
# 문제 1. 이상치 탐지 : 구매금액이 (0 ~ 1,000,000원) 범위를 벗어나면 이상치로 판단한다.

count = 0

for o in orders:
    if o["구매금액"] is not None:
        if o["구매금액"] < 0 or o["구매금액"] > 1000000:
            count = count + 1

print(count)
```

```python
# 문제 2. 이상치 처리 : 이상치는 다음 기준으로 수정하시오.(1,000,000 초과 → 1,000,000 / 0 미만 → 0)

for o in orders:
    if o["구매금액"] is not None:
        if o["구매금액"] > 1000000:
            o["구매금액"] = 1000000

        if o["구매금액"] < 0:
            o["구매금액"] = 0

orders
```

```python
# 문제 3. 결측치 확인 : 구매금액이 결측치(None)인 데이터의 개수를 구하시오.

count = 0

for o in orders:
    if o["구매금액"] is None:
        count = count + 1

print(count)
```

```python
# 문제 4. 결측치 처리 : 결측치를 제외한 구매금액 평균을 계산한 후 결측치를 평균값으로 대체하시오.

total = 0
count = 0

for o in orders:
    if o["구매금액"] is not None:
        total = total + o["구매금액"]
        count = count + 1

avg = int(round(total / count, 1))
print(avg)

for o in orders:
    if o["구매금액"] is None:
        o["구매금액"] = avg

orders
```

```python
# 문제 5. 회원 가입 연수 계산 (특성 추출) : 현재 연도를 기준으로 회원 가입 후 몇 년이 지났는지 계산하여 "가입연수" 컬럼을 추가하시오.

year = 2026

for o in orders:
    join_year = int(o["가입일"][0:4])
    o["가입연수"] = year - join_year

orders
```

```python
# 문제 6. VIP 여부 생성 (특성 추출)

# 구매금액 기준
# 200000원 이상 → VIP
# 200000원 미만 → 일반 으로 "회원등급" 컬럼을 추가하시오.

for o in orders:
    if o["구매금액"] >= 200000:
        o["회원등급"] = "VIP"
    else:
        o["회원등급"] = "일반"
orders
```

```python
# 문제 7. 기술통계 구매금액에 대한

#개수 
#합계 
#평균 
#최대값 
#최소값 

count = 0
total = 0
max_v = -999999999
min_v = 999999999

for o in orders:
    v = o["구매금액"]

    count = count + 1
    total = total + v

    if v > max_v:
        max_v = v

    if v < min_v:
        min_v = v

avg = total / count

print(f"행의수: {count:,} 구매금액합계: {total:,} 평균: {avg:,} 최대: {max_v:,} 최소: {min_v:,}")
```

```python
# 문제 8. 지역별 평균 구매금액 : 지역별 평균 구매금액을 구하시오.

region_sum = {}
region_count = {}

for o in orders:
    r = o["지역"]
    v = o["구매금액"]

    if r not in region_sum:
        region_sum[r] = 0
        region_count[r] = 0

    region_sum[r] = region_sum[r] + v
    region_count[r] = region_count[r] + 1

for r in region_sum:
    print(r, region_sum[r] / region_count[r])
```

```python
# 문제 9. 회원등급별 평균 구매금액 : VIP 회원과 일반 회원의 평균 구매금액을 구하시오.

vip_sum = 0
vip_count = 0

normal_sum = 0
normal_count = 0

for o in orders:
    if o["회원등급"] == "VIP":
        vip_sum = vip_sum + o["구매금액"]
        vip_count = vip_count + 1
    else:
        normal_sum = normal_sum + o["구매금액"]
        normal_count = normal_count + 1

print(f"vip: {vip_sum / vip_count:,}")
print(f"일반: {normal_sum / normal_count:,}")
```

```python
# 문제 10. 가입연수별 그룹 분석 : 가입연수를 기준으로 다음과 같이 그룹을 나누시오.

# 8. 0~2년, 3~5년, 6년 이상 각 그룹의 평균 구매금액을 구하시오.
price1 , price2, price3= 0, 0, 0 # 그룹별 구매금액 누적
count1, count2, count3 = 0, 0, 0 # 그룹별 카운트 누적


for o in orders:
    y = o["가입연수"]

    if y <= 2: # 2년이하
        price1 = price1 + o["구매금액"]
        count1 = count1 + 1

    elif y <= 5:
        price2 = price2 + o["구매금액"]
        count2 = count2 + 1

    else: # 6년이상
        price3 = price3 + o["구매금액"]
        count3 = count3 + 1

print(f"0~2년: {price1 / count1:,}")
print(f"3~5년: {price2 / count2:,}")
print(f"6~년: {price3 / count3:,}")
```

```python
# 심화 문제 11~13 (딕셔너리, 리스트, 반복문, 조건문을 모두 활용)
# 문제 11 : 가장 구매금액이 높은 고객 TOP 3를 출력하시오.
# 11-2) 연평균 구매금액(소수점1자리 반올림)이 높은 구매고객 TOP3?

#9. Top3

result = []

for o in orders:
    result.append(o)

# bubble sort (단순 버전)
for i in range(len(result)):
    for j in range(len(result)):
        if result[i]["구매금액"] > result[j]["구매금액"]:
            temp = result[i]
            result[i] = result[j]
            result[j] = temp

print(result[0])
print(result[1])
print(result[2])
```

```python
# 문제 12 : 지역별 고객 수를 구하시오.

region_count = {}

for o in orders:
    r = o["지역"]

    if r not in region_count:
        region_count[r] = 0

    region_count[r] = region_count[r] + 1

print(region_count)
```

```python
# 문제 13 : VIP 회원 비율을 구하시오.

# 11. VIP 비율
vip = 0
total = 0

for o in orders:
    total = total + 1
    if o["회원등급"] == "VIP":
        vip = vip + 1

print(vip / total)
print(f"{vip / total * 100:.2f}%")
```
