< 파이썬의 조건 표현식과 조건문 > : 조건표현식은 표현식이고, 조건문은 문장이다.

```python
# 1. 조건표현식(삼항 연산자)
# 조건에 따라 하나의 값을 반환한다.
# 변수 = 참일때 값 if 조건 else 거짓일때 값

age = 20
result = "성인" if age >= 19 else "미성년자"
print(result)

print(f"당신은 {'성인' if age >= 19 else '미성년자'} 입니다") # ""안에 문자열은 ''를 사용하자
```

```python
age = 15
result = " "
if age >= 19:
    result = "성인"
else :
    result = "미성년자"

print(f"당신은 {result} 입니다")
```

```python
# 2. 조건문(if 문)

# - 문은 값을 반환하지 않는다.
# - 콜론문자(:)와 들여쓰기를 구문을 사용하여 함께 실행 되는 문들의 묶음인 코드 블록을 표현한다.

age = 20
if age >= 19:
    print("성인입니다.")
```

```python
age = 15

if age >= 19:
    print("성인입니다.")
else:
    print("미성년자입니다.")
```

```python
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
```

```python
age = 20
has_ticket = True

# 비교, 논리 연산
if age >= 19 and has_ticket:
    print("입장 가능")
else:
    print("입장 불가")
```

```python
is_member = False
coupon = True

if is_member or coupon: 
    print("할인 가능")
```

```python
is_student = False

if not is_student:
    print("학생이 아닙니다.")
```

```python
scores = [95, 60, 80, 45]

for score in scores:
    if score >= 70:
        print("합격")
    else:
        print("불합격")
```

```python
numbers = [3, 8, 11, 20, 25]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} : 짝수")
    else:
        print(f"{num} : 홀수")
```
