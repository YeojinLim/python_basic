< 딕셔너리(dict) >

테이블, 데이터프레임같은 2차원 데이터의 한행을 표현할때 많이 사용됨.

- { key : value } 구조, 값을 찾을때 key를 사용해 빠르게 검색함
- 순서 O (Python 3.7 이상) 
- key는 중복 불가, value는 중복 가능 → key는 불변 객체만 사용가능하다.

```python
# 1. dict 자료형 생성, 다루기

# 빈 딕셔너리
d1 = {}
d2 = dict() # 다른 자료구조를 딕셔너리로 변환 할때 주로 사용

print(d1)
print(d2)
```

```python
# {} 사용 , 표현식이기 때문에 : 사용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))
```

```python
# dict() 사용, 대입문 = 사용
person = dict(name="김철수", age=30, city="서울")

print(person)
```

```python
# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

print(data) # 리스트 출력

fruit = dict(data) # 리스트 → 딕셔너리 변환
print(fruit) # 딕셔너리 출력
```

```python
# * 딕셔너리의 키(Key)는 반드시 해시 가능(Hashable) 해야 한다. 그래서 불변(Immutable) 객체는 키로 사용할 수 있지만, 가변(Mutable) 객체는 키로 사용할 수 없다.
# * 해시 가능(Hashable) :고유한 식별자(정수 값)를 가질 수 있고, 이 값이 평생 변하지 않는다

d = {"name": "홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)  
```

```python
# 리스트는 가변 객체 이기때문에 key로 사용할 수 없다
d = {
    [1, 2]: "리스트"
}
print(d) 
```

```python
# 2. dict객체 연산, 메서드, 함수

# 연산

student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제 : del문 사용, 반환값 없음
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
# 메서드

student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys()) # key 출력
print(student.values()) # value 출력
print(student.items()) # 전부 출력

student.update({"age": 30, "score": 90}) # 수정 및 추가
print(student)

age = student.pop("age") # 반환 값을 뽑아내고 삭제
print(age)
print(student)

student.clear() # 전체 삭제
print(student)
```

```python
# 함수

student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}
print(len(student)) # 길이

s = {
    "c": 3,
    "a": 1,
    "b": 2
}

print(sorted(s)) # 정렬, 키 이름을 정렬하여 반환
print(s) # 원본 변화 없음

print(list(s)) # 딕셔너리 → 리스트 변환, 반환값은 키의 리스트
```
