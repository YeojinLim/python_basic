```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []

# 반복문, 조건문
for i in list_nums:
    if i%2 == 0:
        even.append(i)

print(even) 

# 식으로 표현 → : 삭제
even2 = [i for i in list_nums if i%2 == 0] # 리스트
print(even2) 

even3 = {i for i in list_nums if i%2 == 0} # 딕셔너리
print(even3) 

print([i*10 for i in list_nums])
print([i*10 for i in list_nums if i <= 5])
```

```python
names = ["홍길동", "이순신", "이황"]

print([n + "님" for n in names])

print([n + "님" for n in names if n.startswith("이")])
```

```python
names = ["홍길동", "이순신", "김유신", "박혁거세"] # n[:1] → 성, n[1:] → 이름
names2 = ["admin-1234", "guest-0000", "zeroday-7777"] # - 문자를 토큰으로 이용하여 [0],[1] 생성

print(names)
print({n[:1]:n[1:] for n in names})

print("admin-1234". split("-")[0], "admin-1234". split("-")[1]) 
print({n.split("-")[0]:n.split("-")[1] for n in names2})

for n in names2:
    print(n.split("-"))

print({i:p for i , p in (n.split("-") for n in names2)})
```
