```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []

# 반복문, 조건문
for i in list_nums:
    if i%2 == 0:
        even.append(i)

print(even) 

# 식으로 표현 → : 삭제
even2 = [i for i in list_nums if i%2 == 0] # 리스트
print(even2) 

even3 = {i for i in list_nums if i%2 == 0} # 딕셔너리
print(even3) 

```
