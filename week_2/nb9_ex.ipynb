```python
# 문제 1 : CSV 파일을 읽어 data 리스트([{}, {}, {}, ...])를 생성하시오.

import csv

data = []
with open('data_day9.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    data = list(reader)

data
```

```python
# 문제 2 : 학생들의 이름, 국어, 영어 리스트를 각각 저장 후 출력

names = list(map(lambda row : row['이름'], data))
kors = list(map(lambda row : row['국어'], data))
engs = list(map(lambda row : row['영어'], data))

print(names)
print(kors)
print(engs)
```

```python
# 문제 3 : zip()을 이용하여 학생 이름과 국어 점수를 [(), (), ..] 형태로 저장 후 출력

kor_zip = list(zip(names, kors))
print(kor_zip)
```

```python
# 문제 4 : zip()을 이용하여 학생 이름과 영어 점수를 [(), (), ..] 형태로 저장 후 출력

eng_zip = list(zip(names, engs))
print(eng_zip)
```

```python
# 문제 5 : dict(zip())을 이용하여 다음과 같은 딕셔너리(2개 name_kor_dict, name_eng_dict)를 저장 후 출력

name_kor_dict = dict(zip(names, kors))
print(name_kor_dict)

name_eng_dict = dict(zip(names, engs))
print(name_eng_dict)
```

```python
# 문제 6 : 국어 점수가 높은 순으로 정렬하시오. (stored(), 람다 사용)

r6 = sorted(kor_zip, key=lambda x: x[1], reverse=True)
print(r6)
```

```python
# 문제 7 : 영어 점수가 높은 순으로 정렬하시오. (stored(), 람다 사용)

r7 = sorted(eng_zip, key=lambda x: x[1], reverse=True)
print(r7)
```

```python
# 문제 8 : 국어와 영어의 평균을 계산하여 다음과 같은 자료구조를 생성하시오

avg = list(map(lambda n, k, e: (n, (k + e) / 2), names, kors, engs))
print(avg)
```

```python
# 문제 9 :평균 점수가 높은 순으로 정렬하시오.(람다 사용)

r9 = sorted(avg, key=lambda x: x[1], reverse=True)
print(r9)
```

```python
# 문제 10 : 문제8의 결과에서 평균이 90점 이상인 학생만 출력하시오. (filter(), 람다 사용)

r10 = list(filter(lambda x: x[1] >= 90, avg))
print(r10)
```
