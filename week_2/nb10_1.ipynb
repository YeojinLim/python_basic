< 표현식 → 문 → 모듈 → 패키지 → 프로그램 >

문은 파이썬이 실행하는 하나의 명령이고, 이 문은 표현식을 포함할 수 있다.
문들이 모여 하나의 실행가능한 파일이나 모듈(파일)이 된다.
표현식 → 문 → 모듈(.py 파일) → 패키지(폴더) → 프로그램(전체 시스템) 으로 이해하면 된다.

- 프로그램은 모듈로 구성된다.
- 모듈은 문을 포함하고 있다.
- 문은 표현식을 포함하고 있다.
- 표현식은 객체를 생성하고 처리한다.

프로그램(Program)
 ├─ 모듈(Module)
 └─ 패키지(Package)
      ├─ 모듈(Module)
      │    ├─ 문(Statement)
      │    │    └─ 표현식(Expression)
      │    └─ 문(Statement)
      └─ 모듈(Module)

* 표현식(Expression) : 값을 만드는 코드 / 예) 10 + 20
* 문(Statement) : 실행 명령 / 예) x = 10
* 모듈(Module) : 파이썬 파일 / 예) math.py
* 패키지(Package) : 모듈들을 모은 폴더 / 예) mypackage/
* 프로그램(Program) : 여러 모듈과 패키지로 구성된 전체 애플리케이션 / 예) 쇼핑몰, 웹서비스 등

* 표현식에는 함수호출식, 삼항연산자(조건표현식), 컴프리헨션(반복표현식), 람다(함수표현식)도 있다.

< 모듈과 패키지 >

- 모듈 : py파일
- 패키지 : 동일 업무처리에 사용되는 모듈들을 모아놓은 폴더

1. 사용자 정의 모듈

- 모듈 import
모듈안의 변수, 함수, 클래스등을 사용하기 위해서는 import 키워드를 사용한다.

* import 방법 3가지

1. math 모듈 전체를 가져온다. 사용할 때 반드시 math. 를 붙여야 한다.
import math

2. math 모듈을 m이라는 별명(alias)으로 사용한다.
import math as m

3. math 모듈에서 sqrt 함수만 가져온다. math. 없이 바로 사용 가능.
from math import sqrt

```python
# 패키지는 폴더, 모듈은 파일로 되어있다. 
# "."연산자를 이용하여 단계별로 접근하면 된다

# 모듈 연습
import my_math
print(my_math.add(2,3))
```

```python
from my_math import add as ad

print(ad(2, 4))
```

```python
import my_math as mm
print(mm.add(2,3))

from my_math import add
print(add(3,6))

from my_math import add as a
print(a(1,6))
```

```python
# 사용자 정의 패키지 
# greet 폴더(패키지) 생성

from greet import ko

ko.ko_greet("홍길동")
```
