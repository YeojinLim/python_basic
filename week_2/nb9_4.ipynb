```python
# 컬렉션을 다룰 때 자주 사용되는 내장함수들(고차함수 제외)
# 1. enumerate() : 인덱스와 요소 묶기

names = ["홍길동", "이순신", "강감찬"]

for i, name in enumerate(names):
    print(i, name)
```

```python
# 2. reversed() : 역순 반복

nums = [1, 2, 3]
print(list(reversed(nums)))
```

```python
# 3. any() : 하나라도 True인지 검사

nums = [0, 0, 3]
print(any(nums))
```

```python
# 4. all() : 모두 True인지 검사

nums = [1, 2, 3]
print(all(nums))
```

```python
# 5. zip() : 여러 개의 시퀀스 객체를 같은 위치(인덱스)의 값끼리 묶어주는 함수이다.

# 1) zip -> list

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
result = zip(names, scores)
print(list(result)) # [('홍길동', 90), ('이순신', 80), ('김유신', 100)]

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
grades = ["A", "B", "A"]

print(list(zip(names, scores, grades)))
```

```python
# 반복문에서 사용

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

for name, score in zip(names, scores):
    print(name, score)

# 홍길동 90
# 이순신 80
# 김유신 100
```

```python
# 2) zip -> dict

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = dict(zip(names, scores))

print(result) # {'홍길동': 90, '이순신': 80, '김유신': 100}
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = {name: score for name, score in zip(names, scores)}

print(result)
```
