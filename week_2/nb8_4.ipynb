- 분석 결과로 고객이 "어떤 숫자가 필요한지" 모두 지정해주는 경우는 없다. 분석가가 판단해서 적절한 기술통계를 선택해야 한다. 
- csv파일를 분석 후 적절한 기술통계를 선택해 보자.

```python
# csv파일 읽기

import csv

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

data
```

```python
# 결측치 처리 1. 나이 : 평균 값을 구해 결측치 처리

ages = []

for row in data:
    if row["나이"] != "":
        ages.append(int(row["나이"]))

# 평균 나이 계산 (소수점 버림 처리)
avg_age = sum(ages) // len(ages) if ages else 0
print(f"평균 나이: {avg_age}세\n")
```

```python
for row in data:
    # 1. 나이 처리
    if row["나이"] == "":
        row["나이"] = avg_age
    else:
        row["나이"] = int(row["나이"])  # 정상적인 나이는 int로 변환

data
```

```python
# 결측치 및 이상치 처리 2. 국어, 영어 점수 : 빈칸, 문자, 마이너스 값은 0점 처리, 100 초과 값은 100점 처리

for row in data:
    for subject in ["국어", "영어"]:
        val = row[subject]
        
        # 글자나 빈칸, NaN인 경우 0점 처리
        try:
            score = int(val)
        except ValueError:
            score = 0  
            
        # 마이너스 값은 0점 처리, 100 초과 값은 100점 처리
        if score < 0:
            score = 0   # 마이너스나 0은 0점 처리
        elif score > 100:
            score = 100 # 100 초과는 100점 처리
    
        row[subject] = score

data
```

```python
# 평균값 구하기

kor = []
eng = []
total = []

for row in data:
    kor.append(row["국어"])
    eng.append(row["영어"])


kor_avg = sum(kor) / len(kor)
eng_avg = sum(eng) / len(eng)
total_avg = (sum(kor) + sum(eng)) / (len(kor) + len(eng))

print(f"국어 평균 점수: {kor_avg:.2f}점")
print(f"영어 평균 점수: {eng_avg:.2f}점")
print(f"총 평균 점수: {total_avg:.2f}점")
```

```python
# 합격/불합격 판단

pnp = []

for row in data:
    # 1. 개인별 평균 점수 구하기 (국어와 영어는 앞 단계에서 이미 숫자로 바뀐 상태!)
    total_avg = row["국어"] + row["영어"]
    personal_avg = (row["국어"] + row["영어"]) / 2
    
    # 2. 60점을 기준으로 합격/불합격 판정하기
    if personal_avg >= 60:
        result = "합격"
    else:
        result = "불합격"
        
    # 3. 딕셔너리에 계산된 평균과 결과 추가하기
    row["총점"] = total_avg
    row["평균"] = personal_avg
    row["결과"] = result
    
    # 4. 완성된 딕셔너리를 결과 리스트에 차곡차곡 넣기
    pnp.append(row)

data
```

```python
# 순위 구하기

# *순위 정렬해서 순차적 번호를 붙이면 되지만 사전타입 리스트를 정렬하려면 람다가 필요해서 여기선 간단한 알고리즘으로 순위 구현
# for중첩 : 전체학생의 순위를 구하기 위해 반복하고, 그 안에서 전체 학생의 총점과 비교를 위해 중첩반복


for current_student in data: # 모든 행에 반복문으로 순위를 구함
    
    # 한행을 가져오면 기본값으로 현재행(학생) 순위는 1위
    rank = 1

    # 현재행(학생)의 총점을 반복문으로 모든행들의 총점을 비교하여 높은 학생이 있을때 마다 1증가
    # 한 학생의 순위는 나보다 총점이 높은 학생의 수 +1 하면 그 학생의 순위가 나온다
    for other_student in data: # 전체를 반복
        if other_student["총점"] > current_student["총점"]:
            rank += 1

    # 순위 = 나보다 높은 학생 수 + 1
    current_student["순위"] = rank

data
```
