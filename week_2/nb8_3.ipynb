< with문과 파일(txt, csv) 입출력 >

1. txt 파일
- 목적 : 일반 텍스트 저장
- 형식 : 자유롭게 작성 가능
- 구분자 : 없음(자유 형식) 
- 파싱 : 직접 [split()]필요
- 사용 예 : 메모, 로그, 설정 파일

2. csv 파일

- 목적 : 표 형태의 데이터 저장
- 형식 : 행과 열 구조
- 구분자 : 보통 , (쉼표)
- 파싱 : [csv.reader()]가 자동 처리
- 사용 예 : 엑셀 데이터, 데이터 분석


※ with문

with문은 파일이나 자원(Resource : 프로그램이 시스템에게서 빌려서 사용하는 것) 등을 사용한 뒤 자동으로 정리(닫기)해 주는 문법이다.

with 자원해지가 필요한 객체 as 변수:
    실행할 코드

```python
# with문 사용하지 않는 경우

f = open("students.csv", "r", encoding="utf-8")
print(f.read())
f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다. 
```

```python
# try-except-finally 문으로 자원 해지

try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10/0)
finally:
    f.close()
    print("자원 해지 성공")
```

```python
# with문을 사용하는 경우

with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
```

1. txt파일 파싱

open()함수 : 

- 사용방법
파일변수 = open(file, mode='r', buffering=-1, encoding=None, ...)

encoding : 안에 있는 글자를 어떻게 읽을것인지

- 열기 모드 (mode 인자값)
* r : 읽기 전용
* w : 덮어쓰기
* a : 뒤에 추가
* r+ : 읽기+쓰기
* w+ : 덮어쓰기+읽기
* a+ : 추가+읽기

```python
with open("students.txt", "r", encoding="utf-8") as f:
    # 한 줄씩 읽기.
    for line in f: 
        print(line) # 라인 끝의 줄바꿈(\n)이 문자 포함되어 두번 줄바꿈이 일어난다(print함수도 줄바꿈이 기본값)
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
a = '홍길동,90'.split(",")
print(a[0])
print(a[1])

a, b = '홍길동,90'.split(",")
print(a)
print(b)
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        print(name, int(score))
```

```python
# 개별 리스트 자료구조화

names = []
scores = []

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        names.append(name)
        scores.append(score)

print(names)
print(scores)
```

```python
# 딕셔너리로 자료구조화

students = {}

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

```python
# 하나의 리스트 자료구조화

s =[] # [(), (), (), ...]

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        a = line.strip().split(",")
        s.append(tuple(a))

print(s)
```

2. csv 파일

```python
# * csv파일을 불러와서 사용할때는 open()함수 보다는 csv파일 전용 모듈 사용을 권장

import csv

# open()은 내장함수
with open("students.csv", "r", encoding="utf-8") as f:
    # reader() : csv모듈 안에 있는 함수
    reader = csv.reader(f)
    
    for row in reader:
        print(row)
        
print(type(reader))
print(type(row))
```

```python
# 헤더(첫번째 줄) 제외

import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader) # 헤더 건너뛰고 읽기

    for row in reader:
        print(row)
```

```python
# DictReader()를 사용하여 첫번째 줄을 키로 하는 딕셔너리로 반환

import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인은 키, 한줄의 내용은 값

    for row in reader:
        print(row)
```

```python
# ※ csv모듈의 reader반환 타입은 속도는 읽기 속도는 빠르지만(iterator), 자료구조 내부를 핸들링하는데는 여러가지 제한이 있다. 
# 필요하다면 list(reader)로 리스트로 만들어 인덱싱, 슬라이싱, len() 등을 사용하자.
# 큰 CSV 파일은 reader 반복으로 처리하고, 작은 CSV 파일은 list(reader)로 메모리에 올려 사용하자.

import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)

print(data[2])

data[2]["영어"] = 'None' # csv 원본 파일은 변화 하지 않는다.
print(data[2])
```

```python
# 3. CSV 파일로 저장
# csv.DictReader()로 읽어서 data를 수정했다면, csv.DictWriter() 또는 일반 파일 쓰기를 이용해 다시 저장하면 된다. 

import csv

# students_result.csv 파일을 쓰기(w) 모드로 열기
# newline="" : 빈 줄이 추가되는 현상 방지
# encoding="utf-8-sig" : 한글이 엑셀에서 깨지지 않도록 저장
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
