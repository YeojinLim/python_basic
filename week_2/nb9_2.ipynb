```python
# 1. 반환값이 없는 함수

def hello():
    """
    '안녕'을 출력하고 None 값을 반환하는 함수 
    """
    print("안녕")

hello()

#※ 함수 이름 바로 아래에 작성하는 """ ... """ 문자열은 독스트링(Docstring, Documentation String) 이라고 한다.
# - 독스트링은  함수, 클래스, 모듈의 설명(문서) 을 작성할 때 사용한다. 
# - 일반 주석(#)과 달리 프로그램에서 읽을 수 있다. 
# - help() 함수나 IDE(VSCode, PyCharm 등)에서 자동으로 표시된다.
```

```python
result = hello()

# 함수호출은 표현식이다. 표현식은 값을 남긴다. 반환값이 없는 함수를 호출하면 None값을 남긴다.
print(result) # result에는 None값이 할당되어 있다.

help(hello)
```

```python
# 4. 함수 호출시 "불변객체 인수 vs 가변객체 인수"를 사용했을때 차이

# 1) 불변 객체(Immutable)를 인수로 전달 : int, float, str, tuple, bool, 등 → ((숫자, 문자, 튜플, 참/거짓))

def func(x):
    print("함수 안(변경 전):", x, id(x))
    x += 10
    print("함수 안:", x, id(x)) # id : 메모리 주소

a = 10

func(a)

print("함수 밖:", a, id(a))

# 함수 안: 20 140735354414600
# 함수 밖: 10 140735354414280
# 함수 안의 x만 변경, 함수 밖의 a는 그대로 (불변)
# id()값이 다르다
```

```python
# 2) 가변 객체(Mutable)를 인수로 전달 : list, dict, set 등 → ((리스트, 딕셔너리, 세트))

def func(nums):
    nums.append(4)
    print("함수 안:",nums, id(nums))

data = [1, 2, 3]

func(data)

print("함수 밖:", data, id(data))

# 함수 안: [1, 2, 3, 4] 1732383152768
# 함수 밖: [1, 2, 3, 4] 1732383152768
# id()값이 동일하다.
```

```python
# 함수 코드블록 안에서 새로운 객체를 대입하면 참조가 바뀌면서 인수와는 다른 객체를 참조한다.

def func(nums):
    print("새로운 값 대입 전:", nums, id(nums))
    nums = [100, 200]
    print("새로운 값 대입 후:", nums, id(nums))

data = [1, 2, 3]
func(data)
print(data, id(data))

# 새로운 값 대입 전: [1, 2, 3] 1732382938432
# 새로운 값 대입 후: [100, 200] 1732383079680
# [1, 2, 3] 1732382938432
```

- 가변 객체를 함수에 전달하면 매개변수와 인수가 동일한 객체를 참조한다. 
- 따라서 객체 자체를 수정하면 원본도 함께 변경된다. 
- 반면 새로운 객체를 대입하면 참조가 바뀔 뿐 원본은 변경되지 않는다.
- 파이썬은 값을 복사해서 전달하는 것이 아니라 객체의 참조를 전달한다(Pass by Object Reference)
- 불변 객체는 수정할 수 없어 새로운 객체를 만든다.

5. 함수 코드블록안에서 전역변수의 사용

함수 코드블록 안에 전역변수를 사용하는 것은 좋지 않은 방법이다.  
대부분의 언어들은 함수안에 전역변수를 사용하는 것을 코딩 규칙으로 사용하지 않도록 한다.
왜 문법으로 막지 않느냐? "언어가 모든 것을 금지하기보다, 개발자가 책임 있게 사용하도록 한다"는 것이 프로그래밍 언어의 철학이다.

※ 전역변수는 문법적으로 사용할 수 있지만, 함수가 전역변수에 의존하면 재사용성과 유지보수성이 떨어진다. 따라서 필요한 값을 매개변수로 전달하는 것이 가장 좋은 방법이며, 전역변수는 주로 상수나 프로그램 전체에서 공유하는 설정값의 읽기용도로만 사용하는 것이 좋다.

```python
# # 전역 변수 : 함수 정의와 같은 코드 레벨에 있는 변수

x = 10 # 전역변수 x : 사용하지 않는 편이 좋다.

def fun_global():
    print(x) # 전역변수 x(10) 사용

def fun_local():
    x = 333 
    print(x) # 지역변수 x(333) 사용 


fun_global()
fun_local()
print(x)
```

```python
# 6. 사용자 정의 함수 실습

# 1) 두 시퀀스의 교집합을 구하는 함수

def my_intersect(a, b):
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

print(my_intersect([1, 2, 3], [2, 3, 5]))
print(my_intersect((1, 2, 3), (2, 3, 5)))
print(my_intersect({1, 2, 3}, {2, 3, 5}))
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
def my_intersect(a, b) -> list: # -> list : 최종 반환값을 리스트로 지정한다는 힌트! 강제성 X
    """
    a : 리스트
    b : 리스트
    반환타입 : 리스트
    내용 : 두개의 리스트에 중복되는 값만을 새로운 리스트로 반환
    """
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

a = [1, 2, 3, 4, 5, 6, 7, 8]
b = [5, 6, 7, 10, 12, 13, 14]
r1 = my_intersect(a, b)
print(r1)
help(my_intersect)
```

```python
# 2) 공통 요소 개수 구하기

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"})) # 키 비교
```

```python
# 3) 차집합 구하기

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 4) 합집합 구하기

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# 5) 포함 여부 확인

# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True # return 되면 함수 종료 

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
# 6) 특정 값 개수 세기

def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 7) 중복 제거 : set 안쓰는 경우

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```
