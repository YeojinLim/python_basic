< while/for 반복문과 break, continue, pass, else를 활용 >

- while문은 일반적인 루프를 작성하는 방법을 제공한다.
- for문은 시퀀스 또는 컬렉션 객체 안의 아이템들을 하나씩 순차적으로 처리하기 위해 사용하며, 반복될 때마다 들여쓰기 코드블록을 실행한다.


1. while 반복문

- while문은 헤드라인에 있는 테스트가 참 값으로 평가되는 한 문의 블록을 반복해서 실행한다.
- while문은 테스트가 거짓이 될 때까지 문의 시작으로 돌아오도록 제어가 유지되어 Loop라고 부른다.
- 테스트가 거짓이 되면, 제어는 while 블록 다음 문으로 넘어간다.

```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10 
# 다중 할당
# 내부적으로 오른쪽의 튜플 (0, 10)을 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1 #a += 1
```

2. for

- 정렬된 시퀸스 또는 가변 객체 아의 아이템들을 순차적으로 처리한다.
- 다음과 같이 단계별 처리하고자 하는 객체와 함께 할당 대상을 헤드라인에 명시하고, 반복하고자 하는 문들의 블록이 따라온다.

for 할당변수 in 객체:
    statements
else:
    statements

```python
for a in [5, 10, 15]:
    print(a)

for a, *b, c in  [(1, 2, 3, 4), (5, 6, 7, 8)]: # * : 남아있는 거 다 리스트로 받겠다
    print(a, b, c)
```

```python
for (a, b) in [(1, 2), (10, 20)]:
    print(a)
    print(b)

for a in [(1, 2), (10, 20)]:
    print(a)
```

3. break, continue, pass, loop else

* break : 반복문을 즉시 종료
* continue : 현재 반복만 건너뛰고 다음 반복 수행
* pass : 아무 동작도 하지 않음
* loop else : [break]없이 반복문이 정상 종료되면 실행

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

print("반복 종료")
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i) 

# 1 2 4 5
```

```python
i = 1

# if i % 2 == 0 : # 블록코드없이 이 상태면 에러 발생
# def my_func(): # # 블록코드없이 이 상태면 에러 발생

# 다른 언어같으면 빈 괄호를 이용하여 빈 내용을 표현
# if (i % 2 == 0) {}

# 파이썬은 미리 만들어 둘때 빈 내용에 pass 키워드를 사용
def my_func():
    pass

if i % 2 == 0 :
    pass  # 빈 블럭 표시, pass 없으면 에러 발생

for i in range(3):
    if i == 1:
        pass
    print(i)
```

```python
# 루프 else
# break가 없는 경우

for i in range(3):
    print(i)
else:
    print("정상 종료")
```

```python
# # break가 있는 경우

for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("정상 종료") # break가 실행되었으므로 else는 실행되지 않는다.
```

```python
# while문에서도 동일하게 사용 가능하다

i = 1

while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1

print("반복 종료")
```

```python
i = 0

while i < 5:
    i += 1

    if i == 3:
        continue # 현재 반복만 건너뛰고 다음 반복 수행, 3은 출력 되지 않는다

    print(i)
```

```python
i = 0

while i < 3:
    if i == 1:
        pass
    print(i)
    i += 1
```
