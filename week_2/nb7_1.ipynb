< 컴프리헨션(Comprehension) >

파이썬의 컴프리헨션(Comprehension)은 기존시퀀스객체(자료구조)를 이용하여 추가적인 연산을 통하여 새로운 리스트, 딕셔너리, 집합을 만드는 문법니다.
- 기존의 자료구조 객체를 이용해 조합, 필터링등의 추가적인 연산을 통해 새로운 리스트, 딕셔너리, 집합 객체를 생성하는 경우에 많이사용된다. 
- 컴프리헨션은 "반복문 + append()"를 한 줄로 줄인 문법

[표현식 for 변수 in 반복가능객체]
[표현식 for 변수 in 반복가능객체 if 조건]

예시
[i**2 for i in range(1,6)]  #1~5까지 반복하면서 i**2 값을 리스트에 넣어라

```python
# 1. 리스트 컴프리헨션
# 일반 반복문을 이용하여 기존 자료구조를 이용하여 새로운 리스트 생성

nums = []

for i in range(0,5): # 0 ~ 4(5-1)
    nums.append(i)

print(nums)
```

```python
# 제곱값 만들기
squares = []

for i in range(1, 6):
    squares.append(i ** 2)

print(squares)
```

```python
# 컴프리헨션

numbers = [i for i in range(1, 6)]
print(numbers)

squares = [i ** 2 for i in range(1, 6)]
print(squares)
```

```python
# 조건문 추가 가능

evens = [i for i in range(1, 11) if i % 2 == 0]
print(evens)
```

```python
names = ["kim", "lee", "park"]
result = [name.upper() for name in names]
print(result)
```

```python
students = [
    {"이름": "김철수", "성적": 95},
    {"이름": "이영희", "성적": 88},
    {"이름": "박민수", "성적": 77}
]

names = [s["이름"] for s in students]
print(names)

scores = [s["성적"] for s in students]
print(scores)

# 우수학생 추출
good_students = [
    s["이름"]
    for s in students
    if s["성적"] >= 90
]
print(good_students)
```

```python
# 2. 딕셔너리 컴프리헨션

squares = {
    i: i**2
    for i in range(1, 6)
}

print(squares)
```

```python
# 이름과 길이
names = ["홍길동", "이순신", "김유신"]
result = {name: len(name) for name in names}
print(result) # {'홍길동': 3, '이순신': 3, '김유신': 3}
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {a.split("-")[0]: a.split("-")[1] for a in accounts}

print(result)
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {
    user: password for user, password in (account.split("-") for account in accounts)
} # accounts 결과 [(,),(,),..]를 다시 컴프리헨션

print(result)
```

```python
scores = [95, 80, 67, 100]

result = {
    s: "합격" if s >= 80 else "불합격"
    for s in scores
}

print(result)
```

```python
#키-값 반전
scores = {
    "홍길동": 90,
    "이순신": 80,
    "김유신": 100
}

result = {value: key for key, value in scores.items()}

print(result)
```

```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []

# 반복문, 조건문
for i in list_nums:
    if i%2 == 0:
        even.append(i)

print(even) 

# 식으로 표현 → : 삭제
even2 = [i for i in list_nums if i%2 == 0] # 리스트
print(even2) 

even3 = {i for i in list_nums if i%2 == 0} # 딕셔너리
print(even3) 

print([i*10 for i in list_nums])
print([i*10 for i in list_nums if i <= 5])
```

```python
names = ["홍길동", "이순신", "이황"]

print([n + "님" for n in names])

print([n + "님" for n in names if n.startswith("이")])
```

```python
names = ["홍길동", "이순신", "김유신", "박혁거세"] # n[:1] → 성, n[1:] → 이름
names2 = ["admin-1234", "guest-0000", "zeroday-7777"] # - 문자를 토큰으로 이용하여 [0],[1] 생성

print(names)
print({n[:1]:n[1:] for n in names})

print("admin-1234". split("-")[0], "admin-1234". split("-")[1]) 
print({n.split("-")[0]:n.split("-")[1] for n in names2})

for n in names2:
    print(n.split("-"))

print({i:p for i , p in (n.split("-") for n in names2)})
```
