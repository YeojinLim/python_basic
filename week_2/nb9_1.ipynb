1. 함수

- 함수는 특정 작업을 수행하도록 만든 재사용 가능한 코드의 묶음이다.
- 함수는 프로그램에서 한 번 이상 실행할 수 있도록 여러 문(코드의 묶음)을 그룹화하는 장치로, 이름을 붙여 이름으로 호출되어 실행되는 코드의 묶음이다.
- 함수는 연산을 수행할 수 있고, 입력값을 받을 수 있다. 입력값은 실행될 때마다 다른 입력값을 받을 수 있다.
- 하나의 함수로 입력값을 달리하여 다양한 맥락에서 활용할 수 있다.
- 동일한 코드가 실행되어야 하는 경우 "복사해서 붙여넣기"의 대안으로 코드의 중복을 피할 수 있다.
- "복사해서 붙여넣기"의 경우 작업이 변경되면 여기저기 흩어져 있는 코드들을 일일이 변경해야 하지만, 함수를 사용하면 함수 코드 하나만 변경하면 된다.
- 프로그램 구조에서 정의된 역활 단위를 각각의 함수로 구현하면 코드의 재사용성도 높아진다.


2. 함수의 정의와 호출

함수 정의 : def를 사용하여 함수를 만드는 것
함수 호출 : 함수 이름 뒤에 ()를 붙여 함수를 실행하는 것, 함수호출은 표현식이다.
숫자 두 개를 더하는 기능을 여러 번 사용해야 한다면, 한 번만 함수로 만들어 두고 필요할 때마다 호출하면 된다.
- 함수를 만드는 것을 함수의 정의라고 한다.

```python
def add(a, b):
    return a + b

# def - 함수를 정의하는 키워드
# add - 함수 이름
# (a, b) - 매개변수(Parameter)
# : - 함수 코드블록의 시작
# return - 함수 종료 & 결과(객체)를 호출한 곳으로 반환
```

```python
# 함수의 호출(Call) : 정의된 함수를 실행하는 것을 함수 호출이라고 한다.
# 함수 호출은 표현식이다. 함수를 호출한 코드라인에는 반환값을 남긴다. 반환값이 없으면 None값을 남긴다.

result = add(10, 20)
print(result)

# 10, 20 - 인수(Argument)
# add(10, 20) - 함수 호출
# result 에는 함수 호출식의 반환값을 대입
```

```python
add2 = add

print(add)
print(add2)

print(add(1, 3))
r1 = add(10, 20)
print(r1 * 100)

r2 = add("abc", "xyz")
print(r2)

def test():
    pass

r3 = test()
print(r3)
```

```python
# 3. 함수 코드블록 안의 변수 범위(지역 스코프)
# 파이썬은 함수(def)와 클래스(class)를 제외한 대부분의 코드 블록은 새로운 변수 범위를 만들지 않는다. 
# 따라서 if, for, while, try, with 블록에서 생성한 변수는 블록 밖에서도 사용할 수 있다.

def test():
    x = 100
    print(x) # 지역 변수

test()

print(x)   # NameError : name 'x' is not defined, 함수안에서 만들어진 변수 x는 블록 밖에서 사용 불가
```

```python
# - 다른 문의 변수들은 코드블록안의 변수범위(Scope)를 만들지 않는다.

if True:
    x = 10

print(x)
for i in range(3):
    value = i * 10

print(i)
print(value)
try:
    result = 100 / 10
except:
    pass

print(result)
```
