```python
# 날짜/시간 datetime 모듈

import datetime
import time

now = datetime.datetime.now()
print(now)

print(now.year)
print(now.month)
print(now.day)
```

```python
today = datetime.date.today()
print(today)

print(today.weekday())
weekday = today.weekday()

week = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
print(f"오늘은 {week[weekday]} 입니다.")
```

```python
xmas = datetime.date(2026,12,25)
print(f"크리스마스는 {week[xmas.weekday()]} 입니다")
```

```python
print(xmas-today) # d-day 구하기

print(today + datetime.timedelta(days = 1000)) # 오늘부터 1000일 구하기

print(today - datetime.timedelta(days = 100)) # 100일 전 날짜 구하기

print(today - datetime.date(2026, 6, 19))
```
