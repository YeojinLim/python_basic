```python
# 0. 함수도 객체(값)이다.
# - 함수를 변수에 저장

def hello():
    print("안녕하세요!")

# 함수를 변수에 저장
f = hello

# 변수로 함수 호출
f()
```

```python
# - 함수를 다른 함수의 매개값(인수)로 전달

def hello():
    print("안녕하세요!")

def execute(func):
    func() # 함수호출연산자()를 사용 : func 매개변수 값이 함수

execute(hello)
```

1. 람다함수 문법

함수 정의는 함수선언문 이다 - 함수를 정의하는 문
람다는 함수 표현식 이다 - 람다 표현식의 결과로 함수를 반환한다.

※ 파이썬에는

조건문 / 조건표현식(삼항연산자) 
반복문 / 반복표현식(컴프리핸션)
함수(문) / 람다(함수표현식)    가 있다.

# 람다 자체는 함수(객체)를 반환하고 람다를 호출하면 람다의 반환값이 반환된다.
→ 구조 [ lambda 매개변수: 반환값 ]

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)
```

```python
# 람다를 익명함수라고 부르기도 한다

r3 = (lambda a, b: a + b)(10, 50)
print(r3)
```

```python
def test(fun):
    a = int(input("숫자를 입력하세요: ")) # 2
    return fun(a) + 1000

def square(x):
    return x * x
def square2(x):
    return x ** 3
def square3(x):
    return x * 1000

r1 = test(square)
print(r1) # 1004
r2 = test(square2)
print(r2) # 1008
r3 = test(square3)
print(r3) # 3000
```

```python
# 3. 람다함수 즉시 호출 
# 람다의 결과값이 함수이다. 람다의 결과값에 () 호출연산자를 사용하면 람다가 호출된다.
# 람다함수는 다른함수의 인수(매개값)으로 많이 사용된다

s = [('홍길동', 90), ('이순신', 85), ('강감찬', 100)]
r = sorted(s, key=lambda x: x[1], reverse=True)

# sorted( 정렬할데이터, key=정렬기준, reverse=올림내림차순 ) → reverse=False (기본값): 오름차순 / reverse=True : 내림차순
# lambda x: x[1] → x(리스트의 요소 튜플)를 입력받아 x[1](튜플의 인덱스1값)를 반환 (90, 85, 100)
# reverse=True : 내림차순으로 정리

print(r) # [('강감찬', 100), ('홍길동', 90), ('이순신', 85)] 
```

```python
# ※ 람다도 이름없는 함수(함수를 반환하는 표현식)이다. 람다도 함수다.
# 일반 함수를 사용해도 결과는 동일한다.

def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True) # reverse=True : 내림차순
print(r)

# 람다함수 사용
r2 = sorted(s, key=lambda e: e[1], reverse=False) # reverse=False : 오름차순
print(r2)
```

# 4. 람다를 인자로 받는 내장 함수 중 자주 사용되는 것들

함수(주로 lambda)를 인수로 받는 함수를 고차 함수(Higher-order Function)라 한다.

- map(func, data) : 각 요소를 변환 / 모든 요소 변경됨
- filter(func, data) : 조건 검사 / 조건을 만족하는 요소만 선택
- sorted(data, key=func) : 정렬 기준 제공 / 정렬된 리스트 변환
- min(data, key=func) : 최소값 판단 기준 제공 / 가장 작은 요소 반환
- max(data, key=func) : 최대값 판단 기준 제공 / 가장 큰 요소 반환
- any(iterable) : True가 하나라도 있는지 확인 / True or False 반환
- all(iterable) : 모두 True 인지 확인 / True or False 반환

```python
# 1) map() - 모든 요소를 변환 / map(func, data)

nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = list(map(lambda x: x * 2, nums))

print(result) #[2, 4, 6, 8]
```

```python
# 2) filter() - 조건에 맞는 요소만(True 일때만) 선택 / filter(func, data)

nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x: x % 2 == 0, nums))

print(result)
```

```python
# 3) sorted() - 정렬 기준 지정 / sorted(data, key=func)

names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x))

print(result)
```

```python
# * stored()에 람다를 사용하면 데이터 과학에서 자주 사용되는 [dict, dict, ...] 자료구조를 원하는 키값으로 정렬이 가능하다.

students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# "점수"를 기준으로 내림차순 정렬
result = sorted(students, key=lambda x: x["점수"], reverse=True)
print(result)
```

```python
# "이름"을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["이름"])
print(result)
```

```python
# - 여러 키를 기준으로 정렬도 가능하다

students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
# 4) min() - 기준에 따라 최소값 찾기 / min(data, key=func)

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

```python
# 5) max() - 기준에 따라 최대값 찾기 / max(data, key=func)

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result) # banana
```

```python
# 6) any() - 하나라도 조건을 만족하는지 확인 / any(iterable)

nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환
result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```

```python
# 7) all() - 모두 조건을 만족하는지 확인 / all(iterable)

nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```
