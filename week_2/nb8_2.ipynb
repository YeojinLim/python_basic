```python
< try-except문 >

1. try-except문

- 예외(Exception)가 발생했을 때 프로그램이 종료되지 않고 원하는 처리를 수행할 수 있다.
- 예외는 코드 에러는 아니지만 프로그램 실행 중 사용자 입력값, 실행환경 등에 의해 발생하는 오류이다.

try:
    # 예외가 발생할 수 있는 코드

except 예외종류:
    # 예외 처리

else:
    # 예외가 없을 때 실행 (선택)

finally:
    # 항상 실행 (선택)
```

```python
num = input("숫자를 입력하세요") # 7
print(num * 10) # 7777777777

num = int(input("숫자를 입력하세요")) # 7
print(num * 10) # 70
```

```python
num = int(input("숫자를 입력하세요")) # a (문자) 입력시
print(num * 10) # 에러 발생
```

```python
# try-except문 : try 코드블록에서 예외 발생시 except문의 코드 블록이 실행되고, 프로그램은 종료가 되지 않고 다음을 진행한다. 

try:
    num = int(input("숫자 입력 : ")) ** 2
    print(num)
except:
    print("숫자가 아닙니다.") # 문자 입력시 출력

print("다음 실행문...")
```

```python
# except문에서 원하는 특정 예외만을 처리

try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)

except ValueError:
    print("숫자를 입력하세요.") 

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.") # 0 입력시

print("종료")
```

```python
# 2. finally 사용 : 예외 발생 여부와 관계없이 항상 실행된다.

try:
    print(10 / 2)
except:
    print("오류 발생")
finally:
    print("프로그램 종료")
```

3. else 사용

예외가 없을 때 실행

try
 │
 ├── 예외 발생 ─────► except
 │
 └── 예외 없음 ─────► else
          │
          ▼
       finally (항상 실행)

```python
try:
    num = int(input("숫자 입력: "))
    result = 100 / num

except ValueError:
    print("숫자를 입력해야 합니다.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

else:
    print("계산 결과:", result)

finally:
    print("프로그램 종료")
```

```python
try:
    with open("test.txt", "r", encoding="utf-8") as f:
        text = f.read()

except FileNotFoundError:
    print("파일이 없습니다.")

else:
    print(text)

finally:
    print("파일 읽기 작업 종료")
```
