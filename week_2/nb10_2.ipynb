```python
# 실습) 로또번호 출력 
import random

balls = list(range(1,46)) # 1 ~ 45
print(balls)

for i in range(0, 1000000):
    temp = balls[0]
    r_num = int(r.random() * len(balls)) # 0 ~ 44
    # r_num = math.floor(r.random() * len(balls)) : import math 했을 경우 
    balls[0] = balls[r_num]
    balls[r_num] = temp

print(balls) 
print(f"이번 주 로또 당첨 예측 번호: {balls[:6]}")
```

```python
# 실습) 로또번호 출력 2 - random 모듈인 sample() 함수 사용
import random as r

print(f"이번 주 로또 당첨 예측 번호: {r.sample(range(1,46),6)}")

# 숫자 정렬
lotto = r.sample(range(1,46),6)
lotto.sort()
print(f"이번 주 로또 당첨 예측 번호: {lotto}")
```

```python
for i in range(10):
    lotto = r.sample(range(1,46),6)
    lotto.sort()
    print(f"{i+1}번째 로또 당첨 예측 번호: {lotto}")
```

```python
# 로또 번호 생성 함수 

def my_lotto_create(a):
    for i in range(a):
        lotto = r.sample(range(1,46),6)
        lotto.sort()
        print(f"{i+1}번째 로또 당첨 예측 번호: {lotto}")

my_lotto_create(3)
```

```python
def my_lotto_create(a):
    result = []

    for i in range(a):
        lotto = r.sample(range(1,46),6)
        lotto.sort()
        result.append(tuple(lotto))
    return tuple(result)

result = my_lotto_create(5)
print(result)
```

```python
# 실습 2) 주사위 숫자 비교 
import random

def dice_cast():
    d1 = random.randint(1,6)
    d2 = random.randint(1,6)
    return d1 + d2

# p1_dice = {"name": "Bob", "dice_value" : 0}
# p1_dice["dice_value"] = dice_cast()
p1_dice = dice_cast()
p2_dice = dice_cast()

if p1_dice == p2_dice:
    print("DRAW")
elif p1_dice > p2_dice:
    print(f"player 1 Win, 1p : {p1_dice}, 2p : {p2_dice}")
else:
    print(f"player 2 Win, 1p : {p1_dice}, 2p : {p2_dice}")
```
