< 튜플(tuple) >

리스트와 비슷하지만, 소괄호()를 사용한다.(요소1, 요소2, ...., 요소n), 리스트의 경우 항목을 변경 가능하지만 튜플을 항목을 변경할 수 없다.

```python
t = (1, 5, 10)

print (t)
print (type(t))
print (t[1])

# t[1] = 19 << 에러 발생! 튜플은 불변이기 때문에 수정 불가능. 
# TypeError: 'tuple' object does not support item assignment
```

```python
# 1. 튜플의 패킹(Packing), 언패킹(Unpacking) 

# 패킹
t = 1, 3, 5
print(t)
print (type(t))

# 언패킹
x, y, z = t
print(x)
print(y)
print(z)

# 변수와 항목(요소) 수 불일치 Error
# a, b = t 
# ValueError: too many values to unpack (expected 2)
```

```python
# 패킹, 언패킹 시 * 연산자 → 수 불일치 해결 가능
# 변수앞에 *를 쓰면 나머지를 다 그 변수에 리스트로 저장

# a, b = [1, 2, 3, 4] # 에러 발생 ! too many values to unpack 수 불일치
*a, b = [1, 2, 3, 4] 
print(a)
print(b)

*a, b = (1, 2, 3, 4) # 튜플로 생성해도 리스트로 들어옴
print(a)
print(b)

c, *d = (1, 2, 3, 4)
print(c)
print(d)

e, *f, g = (1, 2, 3, 4)
print(e)
print(f)
print(g)
```

```python
a = [1, 2, 3, 4] 
print(a)
print(*a) # print(h[0], h[1], h[2]) : 언패킹 
```

```python
# 2. 리스트와 튜플간의 자료형 변경

x = (1, 3)
print(x)

y = list(x)
print(y)

z = tuple(y)
print(z)
```

```python
# 3. 튜플 연산

a = (1, 2)
b = (3, 4)

print(a + b) # 나열
print((1, 2) * 3) # 반복
```

```python
t = (1, 2, 3)

print(2 in t) # True
print(5 in t) # False
```

```python
t = (10, 20, 30, 40)

print(t[1])
print(t[1:3])
```

```python
# 4. 튜플 메서드와 관련 내장함수

t = (1, 2, 2, 3)
print(t.count(2)) # 2의 개수 : 2
print(t.index(2)) # 인덱스 2 위치 : 1
```

```python
t = (1, 2, 3)

lst = list(t) # 튜플 → 리스트
lst.append(4) # 리스트 확장

t2 = tuple(lst) # 리스트 → 튜플
print(t2)
```
