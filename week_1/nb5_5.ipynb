< 셋(set) >

* 숫자 객체를 이야기할때 정수, 실수, 분수, 정밀숫자, 행렬 등을 말했는데, 집합을 표현하는 set도 있다.
* 수학용 데이터타입이지만 다른 수학 객체들과는 다르게 중복을 가지지 않는 특성때문에 일반 자료형으로 많이 사용된다.
* 리스트같은 인덱스는 존재하지 않기 대문에 필요하면 리스트로 변환 후 수정하고 다시 set으로 변환한다.
* 컬렉션 타입이 아니다.

set : 중복을 허용하지 않는 자료형, 순서가 없음 (인덱싱 불가)

```python
a = {1, 2, 3, 2}
print(a) # 셋은 중복을 허용하지 않는다.

# print(a[0]) # 에러 발생 ! 셋은 인덱스가 존재하지 않는다. TypeError: 'set' object is not subscriptable
print(list(a)[0])
```

```python
# 1. 셋 집합연산

a = {1, 2, 3}
b = {3, 4, 5}

# 수학용 집합 연산이 가능하다
print(a & b) # 교집합
print(a | b) # 합집합
print(a - b) # 차집합
```

```python
# 2. 셋 내장함수

s = {1, 2, 3}

s.add(4) # 하나 추가
print(s)

s.update([5, 6]) # 여러개 추가
print(s)

s.remove(2) # 값 삭제 (없으면 오류)
print(s)

s.discard(10) # 값 없으면 무시
```

```python
s = {1, 2, 3}

print(len(s))
print(2 in s)

s.clear()
print(s)
```

```python
# 읽기 전용 set
fs = frozenset([1, 2, 3])
print(fs)
# fs.add(4) # 에러 발생 ! 읽기 전용 set 이므로 수정 불가능 AttributeError: 'frozenset' object has no attribute 'add'
```

```python
a = {"1": "admin", "pw" : 1234} # 딕셔너리
print(a, type(a))

b = {3, 4, 5} # 셋
print(b, type(b))

c = [] # 리스트
print(type(c))
```

```python
# {}은 빈 set 이 아닌 빈 dict 자료형이다. 빈셋이 필요하면 set() 함수 사용

d = {}
print(d)
print(type(d))

d2 = dict()
print(type(d2))

s = set()
print(s)
print(type(s))
```
