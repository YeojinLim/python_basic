```python
# 화면 출력 : print

print("hi", "bye", sep=" ") # sep는 디폴트 값이 none
print("hi", "bye", sep="/") 
print("hi", "bye", sep="---") 
print("hi", "bye", sep="\\") 
```

```python
# 포맷 지정 출력

#1) f-string(가장 많이 사용)

name = "Park"
age = 30

# f-string 기본 사용
print(f"이름: {name}, 나이: {age}")  # 변수 값을 문자열 안에 직접 삽입

# { 변수 : 규칙 }
# 천 단위 표현
money = 1234567890
print(f"{money:,}원")

# 소수점 자리수 지정
pi = 3.141592
print(f"{pi:.2f}")  # { 변수 : 규칙 }
# .2f : 소수점 둘째 자리까지 출력 → 3.14

# 자리수(폭) 지정
num = 7
print(f"{num:5}")   # 전체 5칸 확보, 오른쪽 정렬 → '    7'
print(f"{num:05}")  # 빈칸을 0으로 채움 → 00007
```

```python
# 2) format() 함수

# 기본 사용
print("이름: {}, 나이: {}".format("Park", 30))  # {} 순서대로 값 들어감

# 인덱스 사용
print("이름: {0}, 나이: {1}".format("Park", 30))  # 번호로 위치 지정

# 이름 지정
print("이름: {name}, 나이: {age}".format(name="Park", age=30))  
# 변수 이름으로 직접 지정 가능
```

```python
# 3) % 연산자 : 잘 안씀

name = "Park"
age = 30

print("이름: %s, 나이: %d" % (name, age))  
# %s = 문자열, %d = 정수

pi = 3.141592
print("%.2f" % pi)  # 소수점 둘째 자리까지 출력 → 3.14
```

```python
# 4) 문자열 정렬

# 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"{'Python':<10}")  # 왼쪽 정렬
print(f"{'Python':>10}")  # 오른쪽 정렬
print(f"{'Python':^10}")  # 가운데 정렬
```
