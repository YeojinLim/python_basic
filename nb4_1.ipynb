```python
name = "smith" 
#길이
print(len(name))

# 오프셋(인덱싱)
print(name[2]) # i

# 슬라이싱 : 잘린값을 반환, 원본이 잘리는게 아니다. [a,b] > a부터 b-1 까지의 글자만 출력
print(name[1:3]) 

# 반복문
for ch in name:
    print(ch)

# 시퀀스[start:stop:step]
print(name[::-1])
print(name[4:2:-1])
```

```python
#불변성 
"""
파이썬의 내장 객체중 하나의 값만 가지는 객체(숫자, bool)는 그 값을 변경할 수 없다.(immutable : 불변)

a = 10
a = 20
에서 a객체가 가진 10라는 값을 20으로 변경하는 것이 아니라 20이라는 새로운 객체를 메모리에 만들고 a가 20을 참조하게 하는 것이다.
파이썬의 문자열도 변경할 수 없다.(immutable : 불변)
"""
s = "python"
print(s[0])
# s[0] = "C" # error
#※ 파이썬 타입 중 list, dict, set은 가변 타입(mutable)이다.(객체 생성 후 내부 데이터를 변경할 수 있는 객체를 말한다)

nums = [1, 2, 3]
nums[0] = 100
print(nums)

#문자열은 변경할 수 없기 때문에 문자열의 연산은 새로운 문자열을 생성 후 반환한다.​
```

```python
#문자열 작성 방법
print('python')
print("python")

# 여러줄 문자열 : 작은 따옴표, 큰 따옴표는 같은 기능, 줄바꿈이 있는 문자열을 쉽게 만들 수 있다.

text = '''
안녕하세요
파이썬입니다
'''
print(text)

text = """
안녕하세요
파이썬입니다
"""
print(text)

# 이스케이프 문자 : \n(줄바꿈), \t(탭), \\(역슬래시 문자 그대로 출력)
print("Hello\nPython")

# Raw 문자열 : 문자열 안의 이스케이프 문자(\n, \t, \\ 등)를 특별하게 해석하지 않고 그대로 표현
print(r"C:\Users\data")
```

```python
# f-string : 문자열 안에 변수 표현식{변수} 사용
name = "Tom"
print(f"이름은 {name}이다") #문자열 포맷방식
print("이름은" + name + "이다")

age =20
# print("나이는" + age + "입니다") # 에러 발생! 연결 연산자(+)는 타입이 같아야 한다. 문자열과 숫자 함께 사용 불가
print("나이는" + str(age) + "입니다")
print(f"나이는 {age}이다") # 문자열 포맷방식을 사용하면 더 쉽게 표현가능
```

```python
#문자열 변환 내장함수(파이썬 언어에 포함되어 있는 함수)
"""
☆ 함수 vs 메서드

- 함수는 전역에서 정의된 독립적인 객체로 대부분 다른 객체를 매개값(인자)으로 받아 어떤 기능을 하는 객체이다.

- 메서드는 특정객체에 소속된 함수로 멤버함수 or 메서드라 부르며 해당객체에 대해서만 어떤 기능을 하는 함수이다.

"""

year = 2026
month = 6
day = 22
# date = year + "-" + month + "-" + day # 에러 발생! : 정수(int)와 문자열(str)은 바로 연결할 수 없기 때문에 str()로 문자열로 변환해야 한다.

year = 2026
month = 6
day = 22

date = str(year) + "-" + str(month) + "-" + str(day)

print(date)

# 숫자 변환 내장함수 : int

a = "123"
b = "10"

print(a + b) # 12310

result = int(a) + int(b)

print(result)
```

```python
# 문자열 메서드 : 메서드는 특정 객체에 소속되어 있고, 그 객체에만 동작하는 함수이며, 호출 표현식으로 실행
# 메서드 호출 표현식 : 객체.메서드이름(매개값)

s = "python"
print(dir(s)) # dir함수: 인자로 전달된 객체에서 이용 가능한 모든 속성 반환 
print(s.upper()) # 대문자로 변환
print(s) # 대부분의 객체 메서드는 자기 자신을 수정, but 문자열은 불변
```

```python
# 대소문자 변환
text = "Python Programming"

print(text.upper())      # 모두 대문자
print(text.lower())      # 모두 소문자
print(text.capitalize()) # 첫 글자만 대문자
print(text.title())      # 각 단어 첫 글자 대문자

s = "python"
print(s.upper()) # 대문자로 변환
print(s) # 대부분의 객체 메서드는 자기 자신을 수정, but 문자열은 불변
```

```python
# 공백제거
text = "   Python   "

print(text.strip())   # 양쪽 공백 제거
print(text.lstrip())  # 왼쪽 공백 제거
print(text.rstrip())  # 오른쪽 공백 제거
```

```python
#문자열 검색
text = "Python Programming"

print(text.find("Pro")) # find : 문자열의 위치 반환, 없는 경우에는 -1 반환함
print(text.find("Java"))
print(text.startswith("Python")) # startswith(a): a로 시작하는지 판별, True/False 반환, 성이 "김" name.startsw
print(text.endswith("ing")) # endswith(a): a로 끝나는지 판별, True/False 반환, 파일중에 한글 파일만 f.endswith(".hwp")
```

```python
#문자열 분리 : 분리하면 리스트가 된다.

data = "apple,banana,orange"

result = data.split(",") 

print(result) # ['apple', 'banana', 'orange']
```

```python
#문자열 결합

fruits = ["apple", "banana", "orange"]
result = ", ".join(fruits)
print(result)

name = ["a", "b", "c"]
result = "님, ".join(name) 
print(result) #출력값 : a님, b님, c / join은 단어와 단어 사이에만 글자를 끼워넣기 때문에, 구조상 맨 마지막 글자 뒤에는 아무것도 붙지 않음!
```

```python
# 문자열 판별 : 비밀번호 설정 등에 사용

print("123".isdigit())      # 문자열이 숫자로만 구성되어 있는가?
print("123ㄱㄴㄷ".isdigit()) 

print("Python".isalpha())   # 문자열이 영문자(또는 한글 등 문자)로만 구성되어 있는가?
print("Python123".isalpha())

print("Python123".isalnum()) # 문자열이 문자와 숫자로만 구성되어 있는가?
print("Python123#".isalnum()) 
```

```python
# 문자열 채우기 : 잘 사용하지 않음, f-string을 사용하면 더 쉽다.

num = "100"

print(num.zfill(5))    # 전체 길이가 5가 되도록 왼쪽을 0으로 채움
print(num.center(10))  # 전체 길이가 10이 되도록 가운데 정렬
print(num.ljust(10))   # 전체 길이가 10이 되도록 왼쪽 정렬
print(num.rjust(10))   # 전체 길이가 10이 되도록 오른쪽 정렬
```
