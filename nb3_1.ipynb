```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
# 2. 실수 (float) ->> 부동 소수점 수! : 오차를 가진다.
# 소수점 16자리까지만 정확도 유지, 뒤 숫자는 근사치로 표현 (출력값  : 0.30000000000000004)

b = 3.14
print(b)
print(type(b))

print(0.1 + 0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str) > 불변 객체(Immutable)
e = "Python"
print(e)
print(type(e))

print(e[1])

# e[1] = "K" # 에러발생! 불변 객체 수정 불가 TypeError: 'str' object does not support item assignment 
```

```python
# 6. 리스트 (list) / 가변 객체(Mutable)
# 가변 객체 : 수정, 추가, 삭제 모두 자유로움 / 동적 배열이라 메모리를 더 쓰고 속도가 비교적 느림 / 게임 캐릭터의 인벤토리, 오늘 할 일 목록 등에 사용

f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

# 값 수정
f[0] = 10
print(f)
f[1] = "abc"
print(f)

# 값 추가
f.append(4)
print(f)

# 값 삭제
del f[1]
print(f)
```

```python
# 7. 튜플 (tuple) / 불변 객체(Immutable)
# 불변 객체 : 한 번 만들면 절대 변경 불가, 읽기 전용 / 고정된 크기라 메모리를 적게 쓰고 속도가 빠름 / 요일, 위치 데이터 등에 사용

g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

# g[1] = 9 # 에러발생! 불변 객체 수정 불가 >TypeError: 'tuple' object does not support item assignment 
```

```python
# 8. 딕셔너리 (dict, 사전) / 가변 객체(Mutable) - 사전, 리스트
h = {"a": 1}
print(h)
print(type(h))

h = {"id": "admin", "pw": "1234"}
print(h)
print(h["id"])
h["pw"] = "0000"
print(h)

h = {"id": "admin", (1,3,5,6): "1234"} # 튜플이 키(Key)로 쓰이는 경우, 컴퓨터(기계)가 이렇게 만든다.
print(h)
```

```python
# 9. 집합 (set)
i = {1, 2, 3, 2} 
print(i) # 출력값: {1, 2, 3} > 중복 값 삭제됨
print(type(i))

#print(i[0]) #키가 없기때문에 일부분을 출력할수 없다 > TypeError: 'set' object is not subscriptable
```

```python
# 10. 범위 (range)
j = range(5) # range(a): 0부터 (a-1)까지 1씩 증가
print(j)
print(type(j))
print(list(j))

j = range(1,7) #range(a, b): a부터 (b-1)까지 1씩 증가
print(j)
print(list(j))
```

```python
# 11. 바이트 (bytes)
k = b"abc"
print(k)
print(type(k))
```

```python
# 12. 바이트배열 (bytearray)
l = bytearray(b"abc")
print(l)
print(type(l))
```

```python
# 13. 없음 (NoneType) : 값, 기능(메서드) 모두 갖고있지 않음
m = None
print(m)
print(type(m))
```

```python
# 14. 함수 (function) : 기능만 가지고 있다. 입력값 다룸
def hello():
    print("Hello") # Tab을 사용
    print("bye")

print("hi") # 함수와 별개

print(hello) # hello라는 변수가 가진 숨겨진 기능 코드값 출력
print(type(hello))
hello() 
```

```python
# 15. 모듈 (module)
import math

print(math)
print(type(math))
```

```python
# 16. 클래스 (type)
class Person:
    pass

print(Person)
print(type(Person))
```
