```python
a = 3
a = "hello"

#1) 숫자 3을 나타내는 새로운 객체 생성
#2) 변수 a가 아직 존재하지 않으면 생성
#3) 변수 a를 새로운 객체 3으로 연결(대입, 할당, 참조)
#4) 문자열 "hello"를 나타내는 새로운 객체 생성
#5) 변수 a를 객체 "hello"로 연결
```

```python
# 공유 참조 : 같은 객체를 참조
# 1. 불변 객체의 공유 참조

a = 10
b = a

print(a, b) # 10 10
print(a is b)  # True 같은 객체 참조
print(a == b)  # == : 값만 비교

a = 20 # 숫자는 불변 객체이므로 값 변경이 아니라 새로운 20을 만들고 a는 20을 참조 (불변객체의 공유참조 값 변경시 새로운 객체를 생성 후 참조)

print(a, b) 
print(a is b)
```

```python
# 문자열도 불변 객체

s1 = "Python"
s2 = s1

print(s1 is s2) # True
print(s1 == s2) # True

s1 = s1 + "3"

print(s1)
print(s2)
```

```python
s1 = "Python"
s2 = "Pytho"

s2 = s2 + "n" # 먼저 s2 + n 을 나타내는 새로운 객체 생성후, s2와 링크

print(s1 is s2) # False
print(s1 == s2) # True

# 서로 다른 객체를 참조하지만 객체의 값이 동일한 경우
```

```python
# 가변 객체의 공유 참조 : list, dict, set, 사용자 정의 클래스

nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) # True
print(nums1 == nums2) # True

nums1[0] = 999
print(nums1)
print(nums2)

nums1.append(40)

print(nums1) # [10, 20, 30, 40]
print(nums2) # [10, 20, 30, 40]
```

```python
nums1 = [10, 20, 30]
nums2 = [10, 20, 30]

print(nums1 is nums2) # False / why? 내용물이 같을뿐, 다른 주소에 생성된 각각의 개체이기 때문!
print(nums1 == nums2) # True
```

```python
# 리스트
# 가변객체의 경우 값이 변경될 수도 있으니 동일한 값의 객체를 대입하여 변수를 만들어도 각각 다른 객체를 참조한다.

a = [1, 2, 3]
b = a

print(a == b) # True
print(a is b) # True

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b) # True
print(a is b) # False

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```
